### AWS Pricing Overview  
AWS offers flexible pricing models so that users can choose based on their budget and usage patterns. You only pay for what you use, and you can save money with long-term commitments or by using unused capacity.

---

### On-Demand Instances  
- **What it is**: Pay per hour or per second (depending on instance type) with no long-term commitment.  
- **Use case**: Short-term workloads, testing, or unpredictable traffic.  
- **Pricing**: Higher than other options because it's flexible and risk-free.

---

### Reserved Instances (RI)  
- **What it is**: You reserve an instance for a 1- or 3-year term.  
- **Use case**: Predictable workloads that run continuously (e.g., web servers, databases).  
- **Pricing**: Up to 75% cheaper than on-demand.  
- **Types**:  
  - **Standard RI**: Biggest savings, less flexible.  
  - **Convertible RI**: Can change instance family or OS.  

---

### Spot Instances  
- **What it is**: Use unused EC2 capacity at up to 90% discount.  
- **Use case**: Flexible, fault-tolerant tasks like batch jobs, big data, machine learning.  
- **Pricing**: Very low, but AWS can terminate the instance when needed.  

---

### Savings Plans  
- **What it is**: Commit to a consistent amount of usage (e.g., $100/month) for 1 or 3 years.  
- **Use case**: Flexible workloads across different instance types and regions.  
- **Pricing**: Up to 72% savings.  
- **Types**:  
  - **Compute Savings Plan**: Most flexible, works across any region and instance type.  
  - **EC2 Instance Savings Plan**: Less flexible but higher savings.  

---

### Dedicated Host  
- **What it is**: Physical server fully dedicated to your use.  
- **Use case**: Compliance needs, bring-your-own-licenses (BYOL), enterprise applications.  
- **Pricing**: Higher cost but full control over physical server.  

---

### Pricing Comparison Table

| Pricing Option      | Best For                             | Savings Up To | Flexibility      | Notes                             |
|---------------------|---------------------------------------|----------------|------------------|------------------------------------|
| On-Demand           | Short-term, unpredictable workloads   | 0%             | Very Flexible    | No upfront payment required        |
| Reserved Instances  | Long-term, steady-state workloads     | ~75%           | Low/Medium       | 1 or 3-year commitment              |
| Spot Instances      | Fault-tolerant, flexible tasks        | ~90%           | Low              | Can be interrupted anytime         |
| Savings Plans       | Long-term usage, flexible workloads   | ~72%           | High             | Applies across multiple services   |
| Dedicated Host      | Compliance or licensing needs         | Varies         | Medium           | Full control over server           |

---

### Tips to Save Money on AWS
- Use **Spot Instances** for non-critical jobs.
- Use **Reserved Instances** for long-term workloads.
- Use **Savings Plans** for flexible long-term savings.
- Right-size your resources using **AWS Cost Explorer**.
- Use **Auto Scaling** to handle variable traffic efficiently.
- Set **budgets and billing alarms** to avoid surprises.

---

### Final Summary
- **On-Demand**: Best for beginners and unpredictable workloads.
- **Reserved**: Best for known, long-term use.
- **Spot**: Best for cost-saving in flexible jobs.
- **Savings Plans**: Best mix of flexibility and cost savings.
- **Dedicated Host**: Best for licensing and regulatory requirements.
