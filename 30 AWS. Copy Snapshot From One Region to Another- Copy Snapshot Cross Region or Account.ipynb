### Copy Snapshot From One Region to Another – Cross Region/Account Snapshot Copy

AWS allows you to **copy an EBS Snapshot from one region to another** or **from one AWS account to another**. This is useful for **disaster recovery, backups, and data migration**.

---

### Why Copy Snapshots?

- **Disaster recovery**: Keep backups in another region.
- **Migration**: Move data between regions or accounts.
- **Share snapshots** with another AWS account.
- **Create AMIs** in different regions.

---

### Key Features

- Snapshots can be **copied across regions or accounts**.
- You can also **encrypt or re-encrypt** snapshots during copy.
- Works for both **manual and automatic snapshots**.

---

### Step-by-Step: Copy Snapshot to Another Region

#### Step 1: Go to Snapshots
- Open AWS Console → **EC2 > Snapshots**
- Select the snapshot you want to copy

#### Step 2: Copy Snapshot
- Click **Actions > Copy Snapshot**
- Set the **destination region**
- (Optional) Change encryption key or description
- Click **Copy Snapshot**

> ✅ AWS will begin copying the snapshot to the selected region.

---

### Step-by-Step: Copy Snapshot to Another AWS Account

#### Step 1: Modify Permissions
- Go to your snapshot → **Actions > Modify Permissions**
- Add the **AWS Account ID** you want to share it with

#### Step 2: In Target Account
- Login to the target account
- Go to **Snapshots > Private Snapshots**
- Select shared snapshot → **Copy Snapshot** to your own account

---

### Notes on Encryption

- If the snapshot is **encrypted**, you can only copy it:
  - To another region **within the same account**
  - To another account **only if using a shared KMS key**

---

### Use Cases

| Use Case                       | Benefit                                         |
|--------------------------------|-------------------------------------------------|
| Backup in another region       | For regional disaster recovery                  |
| Data migration                 | Move application data to a new region/account   |
| AMI availability in other regions | Launch EC2 from snapshot in a new region    |

---

### Summary

| Feature                     | Details                                         |
|------------------------------|-------------------------------------------------|
| Cross-Region Copy            | Yes                                             |
