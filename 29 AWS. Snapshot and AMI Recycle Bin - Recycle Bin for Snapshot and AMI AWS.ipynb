The **AWS Recycle Bin** allows you to **recover deleted EBS Snapshots and AMIs** within a defined **retention period**. This helps prevent accidental data loss and gives you time to restore important resources.

---

### What is AWS Recycle Bin?

- A feature to **temporarily retain deleted Snapshots and AMIs**.
- You can **restore** them during the retention period.
- Works based on **rules** that define what to retain and for how long.

---

### Key Benefits

- **Protection against accidental deletion**
- **Time window to recover** deleted snapshots or AMIs
- No need to immediately rely on backups

---

### How It Works

| Step                     | Description                                      |
|--------------------------|--------------------------------------------------|
| Define Retention Rule    | Set resource type (Snapshot/AMI) and duration    |
| Delete Resource          | Deleted snapshot/AMI goes to Recycle Bin         |
| Recover or Purge         | Restore within retention time or it is deleted   |

---

### Supported Resources

- **EBS Snapshots**
- **Amazon Machine Images (AMIs)**

### Step-by-Step: Enable Recycle Bin for Snapshots or AMIs

#### Step 1: Go to Recycle Bin
- Open AWS Console â†’ Search **"Recycle Bin"**

#### Step 2: Create a Rule
- Click **Create Retention Rule**
- Select **Snapshot** or **AMI**
- Set **Retention Period** (e.g., 7 days, 30 days)
- Add **Resource Tags** (e.g., Key: Type, Value: Backup)

#### Step 3: Review and Create
- Click **Create Rule**

> âœ… Now, when a matching snapshot or AMI is deleted, it goes to the Recycle Bin instead of being permanently deleted.

---

### Recovering Deleted Resources

1. Go to **Recycle Bin Dashboard**
2. Find your deleted Snapshot or AMI
3. Click **Restore** to recover it

---

### Important Notes

- Recycle Bin **only applies to resources covered by rules**.
- You **pay for storage** while the snapshot/AMI is in the Recycle Bin.
- After the **retention period ends**, the resource is permanently deleted.
