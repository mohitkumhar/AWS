### Create and Access Windows Instance in AWS

You can launch and connect to a **Windows EC2 instance** in AWS using the AWS Management Console. Here's a step-by-step guide to help you do it easily.

---

### Step 1: Log in to AWS Console

- Go to [https://aws.amazon.com](https://aws.amazon.com) and log in to your account.
- Navigate to the **EC2 Dashboard** from the **Services** menu.

---

### Step 2: Launch a New Windows EC2 Instance

1. Click **"Launch Instance"**.
2. Enter a **name** for your instance.
3. Under **Application and OS Images (AMI)**, select **Microsoft Windows Server** (e.g., 2019, 2022).
4. Choose an **instance type** (e.g., `t2.micro` for free tier).
5. Under **Key Pair (login)**:
   - Create a **new key pair** or select an existing one.
   - Download the `.pem` file and keep it safe (you need it to access the instance).
6. Under **Network settings**, allow **RDP (port 3389)**.
7. Keep other settings default or customize as needed.
8. Click **"Launch Instance"**.

---

### Step 3: Wait for Instance to Start

- Go to **Instances > View all instances**.
- Wait until the **Status** becomes `running`.

---

### Step 4: Connect to the Windows Instance

1. Select the instance.
2. Click **"Connect"** > **"RDP Client"** tab.
3. Click **"Get Password"**.
4. Upload your **`.pem` key file**.
5. Click **"Decrypt Password"**.
6. Note the **username** (`Administrator`) and **decrypted password**.

---

### Step 5: Use RDP to Connect

- On Windows: Use **Remote Desktop Connection** app.
- On Linux: Use `remmina` or `xfreerdp`.
- On Mac: Use **Microsoft Remote Desktop** app.

Enter the **public IP address**, **username**, and **password**, then connect.

---

### Step 6: Change Password (Optional but Recommended)

- Once logged in, press `Ctrl + Alt + End` â†’ **Change Password**.
- Set a new password for security.

---

### Security Group Settings Reminder

Ensure your instance's **Security Group** allows RDP:

| Type | Protocol | Port Range | Source         |
|------|----------|------------|----------------|
| RDP  | TCP      | 3389       | Your IP / 0.0.0.0/0 |

---

### Summary

| Step                      | Action                                  |
|---------------------------|------------------------------------------|
| Launch Instance           | Select Windows AMI and t2.micro          |
| Create Key Pair           | Download and save `.pem` file            |
| Allow RDP                 | Open port 3389 in Security Group         |
| Get Password              | Decrypt password using `.pem` file       |
| Connect via RDP           | Use public IP and RDP client             |
| Change Windows Password   | Secure your Windows login                |

---

### Tips
- Never share your `.pem` key or Windows password.
- Always restrict RDP access to your own IP for better security.
- Use **Elastic IP** if you want a fixed public IP for the instance.
