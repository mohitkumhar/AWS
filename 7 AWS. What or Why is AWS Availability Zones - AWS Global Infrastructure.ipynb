## **What is an AWS Availability Zone (AZ)?**  
An **AWS Availability Zone (AZ)** is a **physically separate data center** within an **AWS Region**. Each AZ has **independent power, cooling, and networking** but is connected to other AZs in the same region with **low-latency links**.  

### **Why Do AWS Availability Zones Exist?**  
‚úÖ **High Availability** ‚Äì If one AZ fails, others can take over.  
‚úÖ **Fault Tolerance** ‚Äì AZs are built to prevent system failures.  
‚úÖ **Low Latency** ‚Äì Faster communication between data centers.  
‚úÖ **Scalability** ‚Äì Distributes workloads across multiple locations.
