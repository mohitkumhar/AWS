### AWS AMI – Amazon Machine Image – Create Your Own AMI

An **AMI (Amazon Machine Image)** is a **pre-configured template** that contains everything needed to launch an EC2 instance, including the operating system, application server, and applications.

---

### What is an AMI?

- A **blueprint** for creating EC2 instances.
- Includes:
  - Operating system (OS)
  - Application server (e.g., Apache, Nginx)
  - Custom software or configurations

---

### Why Create Your Own AMI?

- **Faster instance launch** with pre-installed software.
- **Consistency**: Launch multiple identical servers easily.
- **Backup**: Save the exact state of your machine.
- **Disaster Recovery**: Quickly recreate environments if needed.
- **Scaling**: Launch many servers with the same setup during load spikes.

---

### Step-by-Step: How to Create Your Own AMI

#### Step 1: Prepare Your EC2 Instance
- Launch and set up an EC2 instance.
- Install required software and do all configurations.

#### Step 2: Create an AMI from the Instance
- Go to **EC2 Dashboard > Instances**.
- Select your instance → Click **Actions > Image and templates > Create Image**.
- Fill details:
  - Name
  - Description (optional)
- Optionally, select storage settings.

#### Step 3: Create Image
- Click **Create Image**.
- AWS will start creating an AMI (check in **AMIs section**).

---

### What Happens After Creating AMI?

| Task                  | Outcome                                 |
|------------------------|-----------------------------------------|
| Create AMI             | A snapshot of the root volume is created |
| Launch instance from AMI | You can create new instances based on that AMI |
| Save configuration     | All installed applications and settings are saved |

---

### Where to Find Your AMI?

- Go to **EC2 Dashboard > AMIs** (in the left sidebar).
- Your custom AMI will be listed there.

---

### Important Notes

- **Storage Costs**: AMIs are backed by snapshots, so they will incur S3 storage charges.
- **Public/Private**: AMIs are private by default; you can share them with other AWS accounts.
- **Regional**: AMIs are region-specific (copy needed to use in another region).

---

### Best Practices

- Name AMIs clearly with date and version.
- Regularly update AMIs if your software stack changes.
- Test instances launched from AMI before using them in production.
