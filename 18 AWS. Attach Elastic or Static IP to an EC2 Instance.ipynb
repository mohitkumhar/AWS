### Attach Elastic (Static) IP to an EC2 Instance

Elastic IP in AWS is a **static, public IPv4 address** that you can assign to your EC2 instance. It allows your instance to have the same public IP even after a reboot or stop/start cycle.

---

### Why Use Elastic IP?

- **Static IP**: Remains the same even if the instance is restarted.
- **Reassociation**: Can be easily moved to another instance.
- **Useful For**: Web servers, DNS mapping, SSH access with a permanent IP.

---

### Steps to Allocate and Attach an Elastic IP

#### Step 1: Allocate a New Elastic IP
1. Go to **AWS Console > EC2 > Network & Security > Elastic IPs**.
2. Click **"Allocate Elastic IP address"**.
3. Choose:
   - **Network Border Group** (same as instance’s region)
   - Click **"Allocate"**

---

#### Step 2: Associate Elastic IP with EC2 Instance
1. Select the newly created Elastic IP.
2. Click **"Actions > Associate Elastic IP address"**.
3. Choose:
   - **Instance**: Select the running EC2 instance.
   - **Private IP**: Select the default or existing private IP.
4. Click **"Associate"**.

---

### Step 3: Verify the Elastic IP
- Go to **EC2 > Instances**.
- Select your instance and check **Public IPv4 address**.
- It should now show the **Elastic IP** you attached.

---

### Releasing an Elastic IP (Optional)
If you no longer need the IP:
1. Disassociate the Elastic IP from the instance.
2. Then select the IP in the Elastic IPs section and click **"Release"**.

> ⚠️ **Note**: AWS charges for unused (unattached) Elastic IPs.

---

### Summary Table

| Task                     | Action in Console                                      |
|--------------------------|--------------------------------------------------------|
| Allocate Elastic IP      | EC2 > Elastic IPs > Allocate Elastic IP               |
| Attach to Instance       | Actions > Associate Elastic IP                        |
| Check Elastic IP         | Instance details > Public IPv4 address                |
| Release IP (if unused)   | Disassociate + Release in Elastic IPs section         |

---

### Important Tips
- One Elastic IP per instance is **free** (as long as it is attached).
- You can **move** the Elastic IP to another instance easily.
- Best used when your application/server needs a **fixed public IP**.
