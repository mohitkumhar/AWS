An **EBS Snapshot** is a way to back up your EBS volume safely. It captures the **entire state of your volume at a point in time** and stores it securely in Amazon S3 (managed by AWS).

---

### Why Create a Snapshot?

- To **protect important data** from accidental loss.
- To **restore** your EC2 volume quickly in case of failure.
- To **clone or copy volumes** across regions or accounts.

---

### Step-by-Step: How to Create Your First Snapshot

#### Step 2: Create Snapshot
- Select the volume â†’ Click **Actions > Create Snapshot**

---

### Important Notes

- Snapshots are **incremental**: after the first full snapshot, only changes are saved in future ones.
- Stored internally in Amazon S3, **not visible directly** in S3.
- You can **copy snapshots** to other regions for disaster recovery.
- You can **encrypt** snapshots and share them with other AWS accounts.

### Best Practices

- Name snapshots with meaningful descriptions.
- Schedule **regular snapshots** for production volumes.
- Delete **unused snapshots** to save cost.
