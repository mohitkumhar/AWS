An **EBS Snapshot** is a **backup of your EBS volume**. It is stored in **Amazon S3** (internally by AWS) and can be used to **restore data or create new volumes**.

---

### Why Use Snapshots?

- **Backup** your data safely and securely.
- **Restore** data if your EC2 or volume fails.
- **Create AMIs** (Amazon Machine Images) from snapshots.
- **Copy** snapshots across regions for disaster recovery.

---

### Key Features of EBS Snapshots

- **Point-in-time** backup of your EBS volume.
- **Incremental**: Only changes since last snapshot are saved.
- Stored in S3 (but not directly accessible like normal S3 files).
- Can be **automated** using Data Lifecycle Manager (DLM).

---

### How to Create a Snapshot (Manual)

1. Go to **EC2 > Volumes**
2. Select the EBS volume you want to back up.
3. Click **Actions > Create Snapshot**
4. Add a name and description.
5. Click **Create Snapshot**

---

### Restore Volume from Snapshot

1. Go to **Snapshots**
2. Select the snapshot
3. Click **Actions > Create Volume**
4. Choose AZ, type, size > Create
5. Attach the new volume to an EC2 instance

---

### Automate Snapshot Backups

Use **Amazon Data Lifecycle Manager (DLM)**:
- Define **policies** for automatic snapshots.
- Set schedule, retention period, and tags.

---

### Snapshot Pricing

- You pay for the **storage used**.
- Since snapshots are **incremental**, cost is reduced.
- Example: If you snapshot a 100 GB volume and only 5 GB changed, only 5 GB is stored.

---

### Common Use Cases

| Use Case                      | Description                                         |
|------------------------------|-----------------------------------------------------|
| Backup important data         | Regular snapshots of databases or apps             |
| Disaster recovery             | Copy snapshots to other regions                    |
| Launch new instances          | Create AMIs using snapshots                        |
| Restore accidentally deleted files | Mount snapshot as volume, recover files       |

---

### Important Notes

- Snapshots are **region-specific** (but can be copied across regions).
- You can **share** snapshots with other AWS accounts.
- Snapshots can be used to create **encrypted volumes**.
- Snapshots of **root volumes** are used in AMIs.

---

### Summary

| Feature            | Details                                      |

| Backup Type        | Point-in-time, incremental                   |
| Storage            | Internally stored in Amazon S3               |
