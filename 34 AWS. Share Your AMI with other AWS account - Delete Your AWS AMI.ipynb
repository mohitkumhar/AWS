### Why Share an AMI?

- **Team collaboration**: Share pre-configured server setups.
- **Multi-account management**: Deploy the same environment in different AWS accounts.
- **Backup and disaster recovery**: Use AMIs across different accounts.

---

### Step-by-Step: Share Your AMI with Another AWS Account

#### Step 1: Make Sure AMI is Unencrypted
- Only **unencrypted AMIs** (or AMIs with customer-managed keys) can be shared easily.
- If encrypted, you need to **share the KMS key** too.

#### Step 2: Go to AMIs Section
- Open AWS Console â†’ **EC2 Dashboard > AMIs**.
- Select your AMI.

#### Step 3: Modify Permissions
- Click **Actions > Modify Image Permissions**.
- Choose **Private** sharing.
- Add the **AWS Account ID** you want to share with.

#### Step 4: Save Changes
- The target account can now see and use the shared AMI under **Private Images**.

> âœ… AMI is now shared with the selected AWS account.

---

### Important Notes for Sharing

- Sharing an AMI **does not share associated snapshots** if encrypted.
- You must **share the snapshot separately** if required.
- If the AMI is public, anyone can access it.

---

### Step-by-Step: How to Delete (Deregister) an AMI

#### Step 1: Go to AMIs Section
- Open **EC2 Dashboard > AMIs**.
- Find the AMI you want to delete.

#### Step 2: Deregister the AMI
- Select the AMI â†’ **Actions > Deregister**.
- Confirm the deregistration.

> ðŸš¨ Deregistering an AMI does **not delete the snapshots** automatically.

#### Step 3: Delete the Associated Snapshots (Optional)
- Go to **Snapshots** section.
- Find and delete the snapshots used by the AMI.
- This will free up storage and stop billing.

---

### What Happens After You Deregister an AMI?

| Action               | Result                                   |
|-----------------------|------------------------------------------|
| Deregister AMI        | Cannot launch new instances from it     |
| Existing Instances    | Existing instances created from that AMI continue to run |
| Storage Cost          | Snapshot storage still charged unless manually deleted |

---

### Best Practices

- Share only required AMIs with trusted accounts.
- Always deregister unused AMIs to avoid confusion.
- Delete snapshots after deregistering to save costs.
