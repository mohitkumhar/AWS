### Encrypt the EBS Volume – What Happens When You Encrypt an EBS Volume

EBS volume encryption helps secure your data by using **AWS-managed or customer-managed keys (KMS)**. When an EBS volume is encrypted, **data at rest, data in transit, and all volume backups (snapshots)** are also encrypted automatically.

---

### What is EBS Encryption?

- Encrypts the **data stored on the volume**.
- Automatically encrypts:
  - **All data at rest**
  - **Data in transit between EC2 and EBS**
  - **All snapshots** and **volumes created from snapshots**

---

### Key Features

- **Uses AWS KMS (Key Management Service)**.
- No application changes needed — works transparently.
- Encryption is applied **at the hardware level** by AWS.

---

### What Happens After Encryption?

| Action                              | Result                                              |
|-------------------------------------|-----------------------------------------------------|
| Store data on volume                | Data is automatically encrypted                     |
| Create snapshot from encrypted vol  | Snapshot is encrypted                               |
| Create volume from encrypted snap   | New volume is also encrypted                        |
| Copy encrypted snapshot             | Must use KMS key in destination region              |
| Share encrypted snapshot            | Only possible if KMS key is shared                  |

---

### How to Encrypt an EBS Volume

#### Method 1: While Launching an Instance

1. Go to **EC2 Dashboard > Launch Instance**
2. In storage section → Click **Edit EBS Settings**
3. Enable **Encryption** and choose a KMS key (or use default)

#### Method 2: Copy Unencrypted Snapshot

1. Go to **Snapshots**
2. Select unencrypted snapshot → **Actions > Copy Snapshot**
3. In the copy settings, enable **Encryption**
4. Create a new volume from the encrypted snapshot

---

### Benefits of EBS Encryption

- **Data security**: Protect sensitive workloads
- **Compliance**: Helps meet regulatory standards (e.g., HIPAA, PCI-DSS)
- **Simple management**: No manual steps after enabling
- **No performance impact**: AWS optimized encryption at hardware level

---

### Limitations

- You **cannot directly encrypt an existing volume**.
- You need to **create a snapshot and then copy it with encryption**, then create a new volume from that snapshot.
- **Encrypted snapshots cannot be shared publicly**.
- You **must use KMS keys** when copying or sharing across accounts/regions.
