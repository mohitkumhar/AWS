### Access Windows EC2 Instance from a Linux Machine Using Remmina

You can use **Remmina**, a remote desktop client for Linux, to connect to your **Windows EC2 instance** using RDP (Remote Desktop Protocol).

---

### Prerequisites
- A running **Windows EC2 instance** with public IP or DNS.
- A **.pem key file** to decrypt the Windows password.
- **Remmina installed** on your Linux system.
- Windows instance **Security Group** allows RDP (port 3389).

---

### Step-by-Step Guide

#### Step 1: Install Remmina
**For Ubuntu/Debian-based systems:**
```bash
sudo apt update
sudo apt install remmina -y
```

---

#### Step 2: Get Windows Administrator Password
1. Go to **AWS Console > EC2 > Instances**.
2. Select your Windows instance.
3. Click **"Connect" > "RDP Client" > "Get Password"**.
4. Upload your `.pem` key file.
5. Click **"Decrypt Password"**.
6. Copy the username (`Administrator`) and decrypted password.

---

#### Step 3: Allow RDP in Security Group
Ensure that the instance's security group has an inbound rule like:

- **Type**: RDP  
- **Protocol**: TCP  
- **Port Range**: 3389  
- **Source**: Your Linux public IP (or `0.0.0.0/0` for testing â€“ not secure)

---

#### Step 4: Open Remmina and Create a Connection
1. Open **Remmina** from your application menu.
2. Click the **"+" icon** (New Connection Profile).
3. Fill in the details:
   - **Protocol**: RDP - Remote Desktop Protocol
   - **Server**: `<Public IP of Windows EC2>`
   - **Username**: `Administrator`
   - **Password**: Paste the decrypted password
   - **Name**: (Any name you like for the connection)
4. Click **Save and Connect**.

---

### Optional: Quick Connect Without Saving
1. Open Remmina.
2. In the top bar, choose:
   - **Protocol**: RDP
   - **Server**: `<Public IP>`
3. Click **Connect**.
4. Enter:
   - **Username**: Administrator
   - **Password**: (Decrypted password)
5. Hit **OK** to connect.

---

### Troubleshooting
- If it fails to connect:
  - Check if RDP port (3389) is open in **Security Group**.
  - Confirm instance is in **Running** state.
  - Verify correct **username/password**.
  - Try restarting the instance if needed.
