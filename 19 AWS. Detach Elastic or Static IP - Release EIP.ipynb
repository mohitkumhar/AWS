### Detach and Release Elastic (Static) IP in AWS

Elastic IPs (EIP) are public static IP addresses provided by AWS. If you no longer need an EIP, you should **detach it from the EC2 instance** and **release it** to avoid unnecessary charges.

---

### Why Detach and Release Elastic IP?

- **Cost Saving**: AWS charges for unused (unattached) Elastic IPs.
- **Clean Resource Management**: Prevents resource wastage.
- **Free Up IPs**: Makes IPs available for other instances or users.

---

### Step 1: Disassociate (Detach) Elastic IP from EC2

1. Go to **AWS Console > EC2 > Elastic IPs**.
2. Select the **Elastic IP** you want to detach.
3. Click **"Actions" > "Disassociate Elastic IP address"**.
4. Confirm the disassociation.

---

### Step 2: Release the Elastic IP

1. After disassociation, the IP becomes "Not associated".
2. Select the same Elastic IP.
3. Click **"Actions" > "Release Elastic IP address"**.
4. Confirm the release.

> ⚠️ **Warning**: Once released, you may not get the same IP again.

---

### When Should You Release an EIP?

- When the instance using it is terminated or no longer needs a static IP.
- To avoid **charges** for unused Elastic IPs.
- If you're switching to a **different IP allocation**.

---

### Summary Table

| Action                        | Where to Do It                                    |
|-------------------------------|---------------------------------------------------|
| Disassociate Elastic IP       | EC2 > Elastic IPs > Actions > Disassociate        |
| Release Elastic IP            | EC2 > Elastic IPs > Actions > Release             |
| Avoid charges for unused IPs  | Always release EIP if not attached                |

---

### Best Practices

- **Only allocate Elastic IPs when necessary**.
- **Release immediately** when not in use.
- **Use DNS** (like Route 53) instead of hardcoding IPs when possible.

---

### Final Note

- Each AWS account gets a limited number of EIPs per region (default: 5).
- AWS **does not charge** for **one attached** Elastic IP per instance.
- Charges apply if:
  - The IP is **not associated** with any running instance.
  - You have **more than one EIP** attached to the same instance.
