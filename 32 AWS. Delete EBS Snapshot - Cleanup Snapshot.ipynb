### Delete EBS Snapshot – Cleanup Snapshot

Over time, you might have many EBS snapshots that take up space and cost money. **Deleting EBS snapshots** helps you **clean up old backups and save storage costs**.

---

### What is an EBS Snapshot?

- A **backup of an EBS volume** stored in Amazon S3 (managed by AWS).
- Useful for restoring volumes, migrating data, or creating AMIs.
- Snapshots **accumulate over time** if not managed.

---

### Why Delete EBS Snapshots?

- **Save storage costs** (you pay for each GB stored).
- **Remove outdated or unnecessary backups**.
- **Maintain a clean environment** by keeping only needed backups.

---

### Step-by-Step: How to Delete an EBS Snapshot

#### Step 1: Open EC2 Console
- Go to **AWS Management Console** → **EC2 Dashboard** → **Snapshots**.

#### Step 2: Select Snapshot
- Find the snapshot you want to delete.
- Select the checkbox next to it.

#### Step 3: Delete
- Click **Actions > Delete Snapshot**.
- Confirm the deletion.

> ✅ The snapshot will be permanently deleted.

---

### Important Notes Before Deleting

- **Check dependencies**:
  - If an AMI (Amazon Machine Image) is using the snapshot, **deleting it can affect AMI functionality**.
  - If other snapshots are created using this snapshot (incremental), AWS manages the underlying storage automatically.
- **Encrypted Snapshots**:
  - Deleting encrypted snapshots works the same way.

---

### Best Practices for Cleanup

- **Tag snapshots** when creating them (e.g., with dates or project names).
- **Regularly review** old snapshots.
- Use **Data Lifecycle Manager (DLM)** to automate snapshot deletion.
- Keep at least **one recent backup** before deleting.
